```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
class_name = 'Occupancy'
df = pd.read_csv('../data/datatraining.txt', skipinitialspace=True, na_values='?', 
                 keep_default_na=True)

```

```python
df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
df['Time']= df['date'].apply(lambda x:x.time())
df['Date']= df['date'].apply(lambda x:x.date())
df['Time'] =  pd.to_timedelta(str(x) for x in df['Time'])
df['Date'] =  pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.drop('date', axis=1, inplace=True)
```

```python
df['Hour']= df['Time'].apply(lambda x:str(x)[7:9])
df['Day']= df['Date'].apply(lambda x:str(x)[8:10])
```

```python
print(len(df))
print(len(df['Date'].unique()))
```

```python
# Set up a grid of plots
fig, axes = plt.subplots(2, 1, figsize=(10,10))

# Histogram of AgeFill segmented by Survived
df1 = df[df[class_name] == 0]['Day']
df2 = df[df[class_name] == 1]['Day']

axes[0].hist([df1, df2], 
             bins=10, # bin_size
             range=(0,10 ), 
             stacked=True)
axes[0].legend(( 'free','Occupated'), loc='best')
axes[0].set_title('Occupancy by days')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Occupancy')
```

```python
# Set up a grid of plots
fig, axes = plt.subplots(2,1 ,figsize=(10,10))

# Histogram of AgeFill segmented by Survived
df1 = df[df[class_name] == 0]['Hour']
df2 = df[df[class_name] == 1]['Hour']

axes[0].hist([df1, df2], 
             bins=24, # bin_size
             range=(0, 24), 
             stacked=True)
axes[0].legend(( 'free','Occupated'), loc='best')
axes[0].set_title('Occupancy by hours')
axes[0].set_xlabel('Hours')
axes[0].set_ylabel('Occupancy')

```
